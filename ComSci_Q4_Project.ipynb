```python
def table(range1, range2):
    tablerow = []

    for value in range(1, range1 + 1):
        tablerow.append(value)

    for value in range(1, range2 + 1):
        for x in tablerow:
            print(f"{x * value:8}", end="")
        print()


range1 = int(input("Please input a range for your table: "))
range2 = int(input("Please input another range for your table: "))

print()
print("Loading Table...")
print()

if range1 and range2:
    table(range1, range2)
else:
    print("ERROR: Invalid Data")
```
